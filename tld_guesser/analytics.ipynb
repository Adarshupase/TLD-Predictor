```python
import pandas as pd

cat_counts = {}

# Read in chunks to count efficiently
for chunk in pd.read_csv("domain_to_tld_cat.csv", chunksize=100_000):
    chunk_counts = chunk["category"].value_counts().to_dict()
    for cat, count in chunk_counts.items():
        cat_counts[cat] = cat_counts.get(cat, 0) + count

# Display final results
for cat, count in sorted(cat_counts.items(), key=lambda x: x[1], reverse=True):
    print(f"{cat:<20} {count:,}")
```
